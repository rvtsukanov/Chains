```python
def running_mean(x, N):
    cumsum = numpy.cumsum(numpy.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
plt.figure(figsize=(20, 10))
plt.grid()
plt.title('Squared reward function')
plt.xlabel('Num iteration')
plt.ylabel('Average reward on trajectory')
plt.plot(running_mean(dqn_data.clip(-5), 100)/100, label='DQN')
plt.plot(running_mean(ac_data, 100)/100, label='Actor-Critic')
plt.plot(running_mean(base_const, 100)/100, label='Constant baseline')
plt.plot(running_mean(base_rand, 100)/100, label='Random baseline')
plt.xlim(0, 10000)
plt.legend()
# plt.ylim(-1, 1)
```
